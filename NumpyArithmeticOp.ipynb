NumPy - Arithmetic Operations

```python
import numpy as np
```

```python
a = np.array([1,2,3,4])
b = np.array([3,4,5,6])
```

```python
c= np.add(a,b)
```

```python
c
```

```python
add = c
```

```python
add
```

```python
sub = np.subtract(a,b)
sub
```

```python
mul = np.multiply(a,b)
div = np.divide(a,b)
power = np.pow(a,b)
modulus = np.mod(a,b)
```

```python
print("Multiplication is ",mul)
print("Division is",div)
print("Power is ",power)
print("Modulus is ",modulus)
```
